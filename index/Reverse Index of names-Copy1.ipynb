```python
columns_for_tokenize = ['matched_name', 'location', 'matched_city', 'matched_address', 'matched_category']

tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True)

sample[columns_for_tokenize] = sample[columns_for_tokenize].fillna("")

for col in columns_for_tokenize:
    sample[col + '_tokens'] = sample[col].apply(tokenizer.tokenize)
```

```python
colums_with_markers = ['matched_title', 'matched_title_2']

def weigh_marker(token_marker_pair):
    new_token = []
    token, marker = token_marker_pair
    if marker not in ['PLACE_MARKER', 'BEG', 'DT'] and len(token) >= 2:
        new_token.extend([token] * 1)
    else:
        new_token.append(token)
    
    return new_token

def weigh_markers(token_marker_pairs):
    if token_marker_pairs == [] or token_marker_pairs is None:
        return []
    new_tokens = []
    
    for token_maker in token_marker_pairs:
        new_tokens.extend(weigh_marker(token_maker))
        
    return new_tokens

for col in colums_with_markers:
    sample[col + '_tokens'] = sample[col].apply(weigh_markers)
```

```python
def get_closest_documents(query, vectorizer, train_corpus_vectors, top_n=10):
    """Vectorizer should be fit on the documents beforehand.
        Returns tuples of (similarity, indexes) of closest documents"""
    # compute similarity to all sentences in the training corpus
    similarities = cosine_similarity(vectorizer.transform([query]), train_corpus_vectors).flatten()
    # get indexes of top n closest sentences
    related_docs_indices = similarities.argsort()[:-top_n-1:-1]
    print(related_docs_indices)
    # return tuples of (similarity score, document id)
    return [(similarities[idx], idx)  for idx in related_docs_indices]
```

```python
def find_closest_word(word, set_of_words):
    suggested_word = ''
    coeff = -1
    if word not in set_of_words:
        for sw in set_of_words:
            fuzz_coeff = fuzz.ratio(sw, word)
            if fuzz_coeff > coeff:
                coeff = fuzz_coeff
                suggested_word = sw
        return suggested_word
    return word
```

```python
def edit_query(query):
    new_query = []
    nq = False
    for w in query.split():
        new_query.append(find_closest_word(w, title_words))
    
    new_query = ' '.join(new_query)
    
    if query != new_query:
        ans = input('Do you want to search for: {} instead?'.format(new_query))
        if ans == 'y':
            return new_query
    
    return query
```

```python
def group_duplicates(tuples, ix, group = []):
    group.append(ix)
    for i, (ix1, ix2) in enumerate(tuples):
        if ix1 == ix:
            del tuples[i]
            group.extend(group_duplicates(tuples, ix2, group))
    
    return group
```

```python
def query_group(df_ix):
    group = cosine_similarity(train_corpus_vectors[df_ix], train_corpus_vectors)[0]
    
    dup_tuples = []
    dub_ids = []

    for j in range(3000):
        if group[j] >= 0.9:
            dup_tuples.append(j)
            
    for ix in dup_tuples:
        dub_ids.append(sample.iloc[ix]['id'])
            
    return dub_ids
```
